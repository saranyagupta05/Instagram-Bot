```python
# For likes 
# liking the first 30 posts of dilsefoodie page 
# printed the message for the post which is already liked 
for i in range(30):
    if i==0:
        first_post = driver.find_element_by_class_name('_9AhH0')
        first_post.click()
        time.sleep(1)
        like_btn = driver.find_element_by_xpath("//span[contains(@class,'fr66n')]/button/*")
        if like_btn.get_attribute("aria-label")=="Like":
            like_btn.click()
            time.sleep(1)
            next_post_click = driver.find_element_by_class_name('_65Bje')
            next_post_click.click()
            time.sleep(3)
        else:
            print("Already liked it")
            next_post_click = driver.find_element_by_class_name('_65Bje')
            next_post_click.click()
            time.sleep(3)
            
    else:
        like_btn = driver.find_element_by_xpath("//span[contains(@class,'fr66n')]/button/*")
        if like_btn.get_attribute("aria-label")=="Like":
            like_btn.click()
            time.sleep(1)
            next_post_click = driver.find_element_by_class_name('_65Bje')
            next_post_click.click()
            time.sleep(3)
            
        else:
            print("Already liked it")
            next_post_click = driver.find_element_by_class_name('_65Bje')
            next_post_click.click()
            time.sleep(3)
            
            
    
# return to dilsefoodie page
driver.find_element_by_class_name('yiMZG').click()
```

```python
## all the followers of “foodtalkindia” that you are following but those who don’t follow you.

## My Following list
profile_icon = driver.find_element_by_partial_link_text('SAMPLE USERNAME')
profile_icon.click()
time.sleep(3)

li = driver.find_elements_by_xpath("//ul[@class='k9GMp ']/li")
following = li[2]
following.click()
time.sleep(3)

following_panel = driver.find_element_by_class_name("isgrP")
my_following_list =[]
current_height_of_following_panel = driver.execute_script("return arguments[0].scrollHeight;",following_panel)
while True:
    ## or => driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight;",following_panel)
    driver.execute_script("arguments[0].scrollTo(0,arguments[1]);",following_panel,current_height_of_following_panel)
    time.sleep(3)
    new_height_of_following_panel = driver.execute_script("return arguments[0].scrollHeight;",following_panel)
    if new_height_of_following_panel == current_height_of_following_panel:
        break
    current_height_of_following_panel = new_height_of_following_panel
    
my_following_list = driver.find_elements_by_xpath("//a[contains(@class,'_0imsa')]")

def extract_text(inpt):
    return inpt.text

series = pd.Series(my_following_list)
my_following_list = series.apply(extract_text)
my_following_set = set(my_following_list)

print("size of my following list = ", len(my_following_list))
for index,username in enumerate(my_following_list,start = 1):
    print(index,"-> ", username)
    
driver.back()
time.sleep(5)

########################################################################################
## My Followers list

li = driver.find_elements_by_xpath("//ul[@class='k9GMp ']/li")
followers = li[1]
followers.click()
time.sleep(3)

followers_panel = driver.find_element_by_class_name("isgrP")
my_followers_list =[]
current_height_of_followers_panel = driver.execute_script("return arguments[0].scrollHeight;",followers_panel)
while True:
    ## or => driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight;",following_panel)
    driver.execute_script("arguments[0].scrollTo(0,arguments[1]);",followers_panel,current_height_of_followers_panel)
    time.sleep(3)
    new_height_of_followers_panel = driver.execute_script("return arguments[0].scrollHeight;",followers_panel)
    if new_height_of_followers_panel == current_height_of_followers_panel:
        break
    current_height_of_followers_panel = new_height_of_followers_panel
    
my_followers_list = driver.find_elements_by_xpath("//a[contains(@class,'_0imsa')]")

def extract_text(inpt):
    return inpt.text

series = pd.Series(my_followers_list)
my_followers_list = series.apply(extract_text)
my_followers_set = set(my_followers_list)

print("size of my followers list = ", len(my_followers_list))
for index,username in enumerate(my_followers_list,start = 1):
    print(index,"-> ", username)
    
driver.back()
driver.back()
```
